```python
import json
```

```python
db = {}
with open('measurements.json') as json_file:
    data = json.load(json_file)
    for key in data.keys():
        db[key] = data[key]
```

```python
config = list(db.keys())[1]
config.split(',')
```

```python
db[config][-1]
```

```python
length = len(db[config])
print(f'Number of measurements: {length}')

relays = len(set([x['relay'] for x in db[config]]))
print(f'Number of unique relays: {relays}')
```
