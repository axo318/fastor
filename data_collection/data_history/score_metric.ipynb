```python
cor_df = df[df[VARIANCE] < 2]
```

## TTLB mean vs variance

## TTLB mean vs std

```python
cor_df = df[df[VARIANCE] < 2]
means = cor_df[MEAN]
std = cor_df[STD]

f, ax = plt.subplots(1, figsize=(8,5))
ax.plot(means, std, '.')
plt.show()
```
